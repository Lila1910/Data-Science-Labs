```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('./course-files/marathon_results_2016.csv', index_col='Bib',
usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)
     
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
     
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
     
df['TotalSeconds'] = df['TotalSeconds'].replace(0,np.nan)
df['HalfSeconds'] = df['HalfSeconds'].replace(0,np.nan)
df.head()
```

```python
df=df.head(10)
group_city=df.groupby('City')
```

```python
for city in group_city:
    print(city[0])
```

```python
for city,city_data in group_city:
    print(city,city_data['Age'].count())
```

```python
for city,city_data in group_city:
    print(city,city_data.loc[city_data['TotalSeconds'].idxmin()])
```

```python
the_best_per_city=pd.DataFrame()
for city,city_data in group_city:
    the_best_per_city=the_best_per_city.append(city_data.loc[city_data['TotalSeconds'].idxmin()])
```
