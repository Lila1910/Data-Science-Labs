```python
import pandas as pd
import numpy as np
fortune=pd.read_csv("./course-files/Fortune_500_2017.csv",usecols=['Rank','Title','Industry','Hqlocation','Sector'])
fortune.head()
```

```python
fortune['Sector']=fortune['Sector'].str.upper()
fortune.head()
```

```python
fortune[fortune['Industry'].str.lower().str.contains('comp')]
```

```python
fortune[['City','State']]=fortune['Hqlocation'].str.split(',',expand=True)
fortune.head()
```

```python
def BuiltShortCut(row):
    industry=row['Industry']
    result=''
    words=industry.split(' ')
    for i in words:
        result+=i[0]
    return result
```

```python
BuiltShortCut({'Industry':'Factory Under Newspaper'})
```

```python
fortune['IndustryShort']=fortune.apply(BuiltShortCut,axis=1)
fortune.head()
```
