```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
marathon = pd.read_csv("./course-files/marathon_results_2017.csv", usecols=["Age","M/F","Country","40K"])
marathon["TimeSeconds"] = marathon["40K"].apply(lambda x: pd.to_timedelta(x,errors='coerce').total_seconds())
marathon.head(5)
```

```python
marathon.plot(kind='scatter',x='Age',y='TimeSeconds')
```

```python
marathon_m=pd.DataFrame(index=marathon.index)
```

```python
men=marathon[marathon['M/F']=='M']
men.head()
```

```python
women=marathon[marathon['M/F']=='F']
women.head()
```

```python
ax=men.plot.scatter(x='Age',y='TimeSeconds',color='Blue',label='Men')
women.plot.scatter(x='Age',y='TimeSeconds',color="Pink",label='Women',ax=ax)
```

```python
women.plot.hexbin(x='Age',y='TimeSeconds',gridsize=30)
```
