```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
df=pd.read_csv("./course-files/marathon_results_2016.csv",index_col='Bib',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df.head()
```

```python
    df['40K'] = df['40K'].replace('-', 0)
    df['Half'] = df['Half'].replace('-', 0)
     
    df['40K'] = df['40K'].apply(pd.to_timedelta)
    df['Half'] = df['Half'].apply(pd.to_timedelta)
     
    df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
    df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
     
    df['TotalSeconds'] = df['TotalSeconds'].replace(0,np.nan)
    df['HalfSeconds'] = df['HalfSeconds'].replace(0,np.nan)
```

```python
df.pivot_table(index='Age',columns='M/F',values='TotalSeconds').head(10)
```

```python
df.pivot_table(index='Age',columns='M/F',values=['TotalSeconds','HalfSeconds']).head(10)
```
