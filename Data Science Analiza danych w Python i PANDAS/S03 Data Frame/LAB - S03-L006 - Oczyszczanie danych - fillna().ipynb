```python
import pandas as pd
import numpy as np
fuel=pd.read_csv("./course-files/fuel.csv",low_memory=False,usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'],index_col='Vehicle ID')
fuel.head()
```

```python
fuel.loc[27705, 'Class'] = np.NaN
fuel.loc[26561, 'Class'] = np.NaN
fuel.loc[27550, 'Fuel Type'] = np.NaN
fuel.loc[27705, 'Combined MPG (FT1)'] = np.NaN
fuel.loc[27681, 'Combined MPG (FT1)'] = np.NaN
fuel.head()
```

```python
fuel.fillna(value=-1).head()
```

```python
replaceRules={'Class':'---','Fuel Type':'---','Combined MPG (FT1)':-1}
replaceRules
```
