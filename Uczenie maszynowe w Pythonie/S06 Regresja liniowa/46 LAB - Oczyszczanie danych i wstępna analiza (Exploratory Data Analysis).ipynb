```python
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    %matplotlib inline    
```

```python
    cols = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS',
            'RAD','TAX','PTRATIO','B','LSTAT','MEDV']
     
    data = pd.read_csv(r"C:\PythonScripts\course-files2\housing.data",
                       sep=' +', engine='python', header=None, 
                       names=cols)
```

```python
data.head()
```

```python
plt.figure()
sns.boxplot(x=data['CHAS'], y=data['MEDV'])
plt.plot()
```

```python
sns.barplot(x=data['CRIM'], y=data['MEDV']) 
```

```python
corr_matrix = np.corrcoef(data.values.T)

fig, ax = plt.subplots(figsize=(11,11))
sns.set(font_scale=1.1)
sns.heatmap(data=corr_matrix,
           square=True,
           cbar=True,
           annot=True,
           fmt='.2f',
           annot_kws={'size':10},
           xticklabels=data.columns,
           yticklabels=data.columns)
```

```python
cols = ['CRIM','INDUS','NOX','RM','AGE','RAD','TAX','PTRATIO','LSTAT','MEDV']
sns.pairplot(data[cols], height=1.5)
plt.show()
```
