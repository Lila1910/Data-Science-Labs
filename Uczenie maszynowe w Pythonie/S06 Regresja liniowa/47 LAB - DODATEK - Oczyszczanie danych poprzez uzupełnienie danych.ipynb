```python
import numpy as np
import pandas as pd
```

```python
diam_org = pd.read_csv(r'C:\PythonScripts\course-files2\diamonds.csv', usecols=['color','price'])
diam_org.head()
```

```python
diam = diam_org.copy()
diam.head()
```

```python
diam_mean_original = diam_org.groupby('color').mean()
diam_mean_original
```

```python
import random
missing_data = random.sample(range(0,diam.shape[0]), 5)
missing_data
```

```python
diam.loc[missing_data]
```

```python
diam.loc[missing_data,'price']=np.NaN
diam.loc[missing_data]
```

```python
diam_mean_start = diam.groupby('color').mean()
diam_mean_start
```

```python
diam_mean_original-diam_mean_start
```
